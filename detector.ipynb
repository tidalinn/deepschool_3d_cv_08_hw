```python
import pandas as pd
import cv2
import os
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('annotations.csv', sep=';', names=['image', 'bbox'])
data.head()
```

```python
os.listdir('data/')
```

```python
file = 'data/'

for name in os.listdir('data/'):
    if data.loc[3, 'image'] in name:
        file += name
```

```python
points = [int(pt) for pt in data.loc[3, 'bbox'].strip('[]').split(', ')]
pt1, pt2 = points[:2], points[2:]
pt1, pt2
```

```python
image = cv2.cvtColor(cv2.imread(file), cv2.COLOR_BGR2RGB)
image = cv2.rectangle(image, pt1, pt2, (255, 0, 0), 2)
plt.imshow(image)
```
